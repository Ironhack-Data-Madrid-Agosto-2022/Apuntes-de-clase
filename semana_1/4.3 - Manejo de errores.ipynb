## `try...except`

Ahora, 쯖칩mo controlamos los errores?

Pues con la sintaxis `try...except`.

El concepto es muy sencillo:

> - `try` bloque de c칩digo que se ejecuta primero.
> > - Si NO existen errores en el bloque `try`, se completar치 ese c칩digo sin ejecutar el bloque `except`
> > - Si existen errores en el bloque `try`, salta al c칩digo del bloque `except`

- Un bloque `try` siempre va acompa침ado de un bloque `except`


- `NOTA:` Un error puede ocurrir en el bloque `except`. En ese caso, el mensaje contendr치 la siguiente advertencia:
````
During handling of the above exception, another exception occurred:
````


`````python
try:
    print(z)

except:
    print('La variable z no existe')

print(90)
`````

`````python
print(z)
print(90)
`````

`````python
try:
    print(z)

except:
    print('La variable z no exite')
    raise ValueError('Compi o me declaras z o me piro')

print(90)
`````

**Truco:** Como puede verse en el mensaje anterior, la l칤nea de c칩digo que contiene el error tiene una flecha al lado. Si el error no est치 en esa l칤nea, estar치 en la inmediatamente anterior. Tambi칠n puede verse el n칰mero de l칤nea.

`````python
import sys
`````

`````python
lst = ['a', 0, 2, 'oe', 3,4,5,6]
`````

`````python
for e in lst:
    try:
        print(F'El elemento es {e}')
        d = 1/int(e)
        
        print(d)
        
    except:
        print('Vaya la has liado', sys.exc_info())
        
        print('siguiente')
        
print('Nos hemos pasao el juego')
`````

### Multiples `except`

Si queremos, podemos tener multiples bloques `except` para manejar diferentes tipos de error por separado. Es algo muy 칰til para manejar como nosotros queramos el flujo de errores, teniendo en cuenta **que solo se ejecutar치 uno de los bloques `except` en caso de surgir un error**.

`````python
try:
    a = 8
    b= 0
    
    print(a/b)
    
except TypeError as e:
    print('estoy en este error')
    print(e)

except ZeroDivisionError:
    print('Como pretendes que divida por 0 alegre')
    
except:
    
    print('deja de liarla tronco')


    
`````

## Programaci칩n defensiva

Todos estos conceptos de manejo de errores nos lleva a la programaci칩n defensiva. El objetivo consiste en tener en cuenta errores sistem치ticos que se pudieran estar cometiendo en nuestro c칩digo.

`````python
clientes = {'a_123': 35.45, 'b_123': 76.45, 'a_asd': 'asda'}
`````

`````python
pasan = []

no_pasan = []
`````

`````python
for k,v in clientes.items():
    
    try:
        total = int(v)*4/3
        pasan.append({k:total})
    except:
        no_pasan.append(k)
`````

`````python
no_pasan
`````

### `finally`

> Los bloques `try` y `except` son los dos bloques comunes en el manejo de errores, pero toda la sintaxis la constituyen cuatro bloques.
El bloque `finally` se ejecutar치 siempre despu칠s del `try` o el `except`. Se ejecutar치 incluso si existe un `return` en otro bloque.

`````python
try:
    raise ValueError
    print('todo va bien')
except:
    print('forzamos el error')
finally:
    print('Siempre me ejecuto 游땕')
`````

`````python
def probar_error():
    try:
        #raise ValueError
        print('todo va bien')
        return 'Todo ok'
    except:
        print('forzamos el error')
        return 'error encontrado'
    finally:
        print('Siempre me ejecuto 游땕')
    
`````

`````python
probar_error()
`````

### `else`

> El bloque `else` puede ser usado como una alternativa al bloque `except`. En ese caso se ejecutar치 antes del bloque `finally`.
